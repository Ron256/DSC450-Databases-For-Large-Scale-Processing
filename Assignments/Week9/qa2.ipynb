```python
import matplotlib.pyplot as plt
from numpy.random import randn
import numpy

fig = plt.figure()                                  # the initial figure is going to be blank

# add a varierty of sub plots to it,
# the first two parameters describe the size of the grid
# that corresponds to the sub figures being added.
# 2 by 2 means we have a grid of four different sub plots
# and we are going to use that to compare and contrast the styles of the different figures. 
# the last parameter refers to which of the sub plots we are adding to.
# so one refers to the top left of the first out of the four subplots in the figure
sp1 = fig.add_subplot(2, 2, 1)


# generate some random numbers
y = randn(30)
print(y)

# Note that we are creating multiple sub plots
# for the time being we are only filling in one of the four elements on the grid
# therefore we only see a signle plot
sp1.plot(y)
fig.set_size_inches(15, 10)
#fig

# creating a nother sub plot and we are using a 2 by 2 grid.
# in this case we are adding the secondor  top-right subplot to the figure 
# you can use different grading within the same figure but that does potentially 
# create overlay which you may or may not want to have in practice. 
# just note that is it possible but we will keep consistent grading into two-by-two
sp2 = fig.add_subplot(2, 2, 2) # so this addes the second sub plot to the top right
# we are going to create several scatter plots of different colors
sp2.scatter(randn(20), randn(20)) # this generates 20 random numbers and another 20 random numbers because we are creating a two dimensional scatterplot
#sp2           # this shows a scatterplot containing 20 different numbers

# we add two more 20-point scatterplot using red and blue
sp2.scatter(randn(20), randn(20), color = 'r') 
sp2.scatter(randn(20), randn(20), color = 'b')
#fig                     # we should see three different scatter plots in the top right corner overlaid together in the same subplot

# next we add a third subplot and we are adding in the botton left, through the subplot
sp3 = fig.add_subplot(2, 2, 3) 
# in this case we are going to create a hsitogram using 40 random numbers and 10 bins and the green color. 
sp3.hist(randn(40), bins=10, color='g')
#fig # As you can see we have three different subplots. one is a line graph, one is a scatter plot of different colors, the other is a histogram
# by default, the histogram binning where we use mean or specific in Numpy mean
# However, you can change the function that is used to compute the bucket in the histogram. 

# lets continue with fourth subplot
# Again, two by two adding to the fourth location over here
sp4 = fig.add_subplot(2, 2, 4)
t = numpy.arange(0, 3.0, 0.01) # I am going to create a range of numbers between zero and three with a step of 100
# so T is a set of 30 numbers incremeting by 0.01
# then I am to compute the cosine function and finally I'm going to plot all of this in the last sum figure. 
s = numpy.cos(2*numpy.pi*t)
sp4.plot(t, s)
fig # This gives us all of the four sobplots
# you can annotations

sp4.annotate('Center', xy = (1.5, 0), xytext = (1.2, 0.5))
#lt.plot(randn(20), color = 'purple', linestyle = 'dashed', marker = 'o')

```
